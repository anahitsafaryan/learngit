# Ձեռագրերի թվայնացման (transcription) ուղեցույց

Այս ուղեցույցի հիմնական նպատակն է 
1.հայալեզու տեղեկության տրամադրումը այն մարդկանց, ովքեր մեր նախագծի շրջանակներում պատրաստվում են զբաղվել ձեռագրերի թվայնացմամբ։ 
2.հայալեզու հակիճ տեղեկության տրամադրումը մեր նախագծի վերաբերյալ՝ Մատթեոս Ուռհայեցու »Ժամանակագրության» առցանց հրատարակությունը։

Ուստի նախ մի փոքր մեր նախագծի մասին։

Մատթեոս Ուռհայեցու «Ժամանակագրությունը կազմվել է 1120-1140 թթ. ընթացքում։ Տեքստն ընդգրկում է շուրջ 180 տարվա պատմություն՝ 951-1128 թթ.։։։։։։։։։։։։; 

Մինչ օրս հայտնի են «Ժամանակագրության» մասնակի կամ ամբողջական տեքստը պարունակող 42 ձեռագիր։ Մեզ հայտնի ամենահին ձեռագիրը, որ պարունակում է «Ժամանակագրության» ամբողջական տեքսը, արտագրվել է 1590-1600 թթ. (Vienna 887): Մնացած բոլոր ձեռագրերը արտագրվել են 17-րդ դ. կամ ավելի ուշ։ Մենք մեր նախագծի շրջանակներում թվայնացնելու ենք շուրջ 10 ձեռագիր, որոնց հիման վրա էլ կունենանք աշխատության առցանց հրատարակությունը։ Ի տարբերություն քննական բնագրերի կազմման ավանդական մեթոդի՝ մենք հիմք տեքստ չենք առանձնացնելու։

Ես այստեղ փորձել եմ ներկայացնել այն ամենը, ինչը սովորել եմ վերջին մի տարում և օգտագործել եմ աշխատանքի ժամանակ։
Մինչ այս ես աշխատել եմ երկու ձեռագրերի վրա՝ Vienna 887 և Matenadaran 1767: Հետևաբար ուղեցույցի հիմքում հենց այս երկու ձեռագրերի թվայնացման արդյունքում հավաքված ինֆորմացիայի ամփոփումն է։ Ինչը նշանակում է, որ այս ուղեցույցը գուցե չկարողանա պատասխանել այլ ձեռագրերի թվայնացման ժամանակ առաջացած որոշ յուրօրինակ հարցերի։ Երբ հանդիպեք նման դեպքի կամ ցանկացած այլ դեպքում, երբ վստահ չեք՝ ինչպես վարվել, պարզապես թողեք մեկնաբանություն (note)

Ուղեցույցի հիմնական աղբյուրներն են Tara Andrews-ի մի քանի աշխատանքներն ու կազմած ուղեցույցը, որ կարող եք գտնել github.com կայքում։
https://github.com/DHUniWien/MatthewEdessa/blob/master/transcription/transcription_guide.md

Թվայնացման գործընթացը հեշտացնելու համար մենք օգտագործում ենք T_PEN (transcription for paleographical and editorial notation) կայքը։ 
http://t-pen.org/TPEN/
Բոլոր այն մարդիկ, ովքեր մեր նախագծի շրջանակներում զբաղվում են թվայնացմամբ, պիտի ունենան հաշիվ (account) այս կայքում։ Այդպիսով բոլորս կկարողանանք իրար հետ կիսվել ձեռագրերով։
Այս կայքում ներբեռնած ձեռագրերի թվայնացման համար նախ պետք է նշել սյունակներն ու տողերը։ Ես ուշադիր կլինեմ, որ ձեզ տրամադրված ձեռագրերն այս առումով արդեն պատրաստի վիճակում լինեն։
